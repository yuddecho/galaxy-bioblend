```python
import yaml
from transmolecule import TransMolecule

with open('config.yaml', 'r') as f:
    config = yaml.safe_load(f)

# 登录
trans_molecule = TransMolecule(config['galaxy_url'], config['api_key'])
```

```python
# 历史记录
# trans_molecule.history.select('f1b9846ab84237e7')
# trans_molecule.history.info()
# trans_molecule.history.content()

trans_molecule.history.create(name="12-4-tteatkkk")

# trans_molecule.dataset.upload(file_path='data/test.json')
trans_molecule.dataset.upload(file_dir='data/pharma')

# 获取数据信息
data = trans_molecule.dataset.get()
print(data)

# trans_molecule.tool.info()
```

```python
trans_molecule.tool.info()
```

```python
# 药效团提取工具
pharma = trans_molecule.tool.get_tool(tool_id="pharma")
pharma_input = pharma.inputs()
print(pharma_input)
```

```python
# 构建输入
pharma_input['pdb_file']['id'] = data['1uko_rec.pdb']
pharma_input['sdf_file']['id'] = data['1uko_ligand.sdf']
print(pharma_input)
```

```python
# 运行工具，提取药效团
job = pharma.run(pharma_input)
print(job)
```

```python
# 多肽设计工具
pharm_pep_gen = trans_molecule.tool.get_tool(tool_name="PharmaPepGen")
pharm_pep_gen_in = pharm_pep_gen.inputs()
print(pharm_pep_gen_in)
```

```python
pharm_pep_gen_in['input_files'][0]['id'] = job['outputs'][0]['id']
pharm_pep_gen_in['num_samples'] = 10
print(pharm_pep_gen_in)
```

```python
job2 = pharm_pep_gen.run(pharm_pep_gen_in)
print(job2)
```
